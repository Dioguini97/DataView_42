```python
players_avg_stats=pd.DataFrame(players_stats.groupby(['Player_ID']).mean()[['FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA',
       'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV', 'PF',
       'PTS']].round(2))
players_avg_stats
```

```python
players_avg_stats.reset_index(inplace=True)
players_avg_stats
```

```python
array_player_name=[]
array_player_team=[]
for ids in players_avg_stats['Player_ID']:
    # print(ids)
    if len(shooting_stats[shooting_stats['PLAYER_ID']==ids]['PLAYER_NAME'].unique()) == 0:
        array_player_name.append(np.nan)
        array_player_team.append(np.nan)
    else:
        array_player_name.append(shooting_stats[shooting_stats['PLAYER_ID']==ids]['PLAYER_NAME'].unique()[0])
        array_player_team.append(shooting_stats[shooting_stats['PLAYER_ID']==ids]['TEAM_NAME'].unique()[0])
```

```python
players_avg_stats['Player_Name']=array_player_name
players_avg_stats['Team_Name']=array_player_team
```

```python
teams_avg_stats=players_avg_stats[['FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM',
       'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV',
       'PF', 'PTS','Team_Name']].groupby('Team_Name').mean().round(2)
teams_avg_stats
```

```python
season_players_avg_stats=players_avg_stats[['FGM', 'FGA', 'FG_PCT', 'FG3M', 'FG3A', 'FG3_PCT', 'FTM',
       'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TOV',
       'PF', 'PTS']].mean().round(2)
season_players_avg_stats
```

Radar Plot

```python
categories = ['FGM', 'FGA', 'FG3M', 'FG3A',  'FTM', 'FTA']
categories_percent = ['FG_PCT' , 'FG3_PCT' , 'FT_PCT']
```

```python
stats_Lebron=players_avg_stats[players_avg_stats['Player_Name']=='LeBron James'][categories].values.flatten().tolist()
stats_Curry=players_avg_stats[players_avg_stats['Player_Name']=='Stephen Curry'][categories].values.flatten().tolist()
```

```python
stats_Lebron_percent=players_avg_stats[players_avg_stats['Player_Name']=='LeBron James'][categories_percent].values.flatten().tolist()
stats_Curry_percent=players_avg_stats[players_avg_stats['Player_Name']=='Stephen Curry'][categories_percent].values.flatten().tolist()
```

```python
stats_lakers=teams_avg_stats[categories].loc['Los Angeles Lakers'].values.flatten().tolist()
```

```python
import plotly.graph_objects as go


fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=stats_Lebron,
      theta=categories,
      fill='toself',
      name='LeBron James'
))
fig.add_trace(go.Scatterpolar(
      r=stats_lakers,
      theta=categories,
      fill='toself',
      name='Team Stats'
))

fig.add_trace(go.Scatterpolar(
      r=season_players_avg_stats[categories],
      theta=categories,
      fill='toself',
      name='Season Stats'
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 22]
    )),
  showlegend=False
)

fig.show()
```

```python
import plotly.graph_objects as go


fig = go.Figure()

fig.add_trace(go.Scatterpolar(
      r=stats_Lebron_percent,
      theta=categories_percent,
      fill='toself',
      name='Product A'
))
fig.add_trace(go.Scatterpolar(
      r=stats_Curry_percent,
      theta=categories_percent,
      fill='toself',
      name='Product B'
))

fig.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 1]
    )),
  showlegend=False
)

fig.show()
```

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# define data set
s1 = pd.Series(['dont have a mortgage',-31.8,'have mortgage',15.65])
s2 = pd.Series(['have utility bill arrears',-21.45,'',0])
s3 = pd.Series(['have interest only mortgage',-19.59,'',0])
s4 = pd.Series(['bank with challenger bank',-19.24,'bank with a traditional bank',32.71])
df = pd.DataFrame([list(s1),list(s2),list(s3),list(s4)], columns = ['label1','value1','label2','value2'])

# create subplots
fig = make_subplots(rows=1, cols=2, specs=[[{}, {}]], shared_xaxes=True,
                    shared_yaxes=True, horizontal_spacing=0)

fig.append_trace(go.Bar(y=df.index, x=df.value1, orientation='h', width=0.4, showlegend=False, marker_color='#4472c4'), 1, 1)
fig.append_trace(go.Bar(y=df.index, x=df.value2, orientation='h', width=0.4, showlegend=False, marker_color='#ed7d31'), 1, 2)
fig.update_yaxes(showticklabels=False) # hide all yticks
```
